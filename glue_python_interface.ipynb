```python
gut_metabolites = gut_metabolites.dropna()
```

```python
# perform PCA on gut_metabolites, extract first 3 principal components, add these as new attributes to gut_metabolites
from sklearn.decomposition import PCA

X = gut_metabolites[metabolites].values
pca = PCA(n_components=3)
pca.fit(X)

print(pca.explained_variance_ratio_, pca.explained_variance_)
```

```python
# add new attributes to gut_metabolites
pca_result = pca.transform(X)

gut_metabolites['PC1'] = pca_result[:,0]
gut_metabolites['PC2'] = pca_result[:,1]
gut_metabolites['PC3'] = pca_result[:,2]

gut_metabolites.to_csv('data/ADMCGUTMETABOLITESLONG_12_13_21.csv', index=False)
```
